```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('./Metal_all_20180116.snappy.parquet')
df_zn = df.loc[df['ligandId'] == 'ZN']
df_zn_single = df_zn.loc[df_zn['interactingChains'] == 1]
```

```python
df
```

```python
len(df)
```

```python
df_clusters = {30:{}, 40:{}, 50:{}, 70:{}, 90:{}, 95:{}, 100:{}}
df_zinc_clusters = {}
for i in range(len(df)):
    if np.isnan(df.loc[i].clusterNumber30):
        continue
    if df.loc[i].clusterNumber30 not in df_clusters[30].keys():
        df_clusters[30][df.loc[i].clusterNumber30] = []
    df_clusters[30][df.loc[i].clusterNumber30].append(i)
    
    if df.loc[i].clusterNumber40 not in df_clusters[40].keys():
        df_clusters[40][df.loc[i].clusterNumber40] = []
    df_clusters[40][df.loc[i].clusterNumber40].append(i)
    
    if df.loc[i].clusterNumber50 not in df_clusters[50].keys():
        df_clusters[50][df.loc[i].clusterNumber50] = []
    df_clusters[50][df.loc[i].clusterNumber50].append(i)
    
    if df.loc[i].clusterNumber70 not in df_clusters[70].keys():
        df_clusters[70][df.loc[i].clusterNumber70] = []
    df_clusters[70][df.loc[i].clusterNumber70].append(i)
    
    if df.loc[i].clusterNumber90 not in df_clusters[90].keys():
        df_clusters[90][df.loc[i].clusterNumber90] = []
    df_clusters[90][df.loc[i].clusterNumber90].append(i)
    
    if df.loc[i].clusterNumber95 not in df_clusters[95].keys():
        df_clusters[95][df.loc[i].clusterNumber95] = []
    df_clusters[95][df.loc[i].clusterNumber95].append(i)
    
    if df.loc[i].clusterNumber100 not in df_clusters[100].keys():
        df_clusters[100][df.loc[i].clusterNumber100] = []
    df_clusters[100][df.loc[i].clusterNumber100].append(i)
```

```python
count = {}
for i in df_clusters.keys():
    count[i] = {}
    for j in df_clusters[i].keys():
        count[i][j] = len(df_clusters[i][j])
```

```python
for i in [30,40,50,70,90,95,100]:
    number = i
    fig = plt.figure(0, figsize=(5, 4), dpi=300)
    ax = fig.add_subplot(111)
    ax.set_title('cluster' +str(number)+' distribution')
#     ax.set_xlim([0, 8000])
#     ax.set_xlim(xmin=0)
    ax.set_xlabel('cluster index')
#     ax.set_ylim([0, 1000])
#     ax.set_ylim(ymin=0)
    ax.set_ylabel('number of sequences')
    l1 = [j for j in count[number].keys()]
    l2 = [j for j in count[number].values()]
    # print (l)
    ax.plot(l2)
#     ax.scatter(l1,l2, s=0.1)
    fig.canvas.draw()
#     plt.show()
    fig.savefig('./dataset_visualization/cluster' + str(number)+ '.png')
    plt.cla()
```
